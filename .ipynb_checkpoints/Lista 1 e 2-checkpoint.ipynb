## Item: Crie 5 séries de preço ajustado para 5 ativos de diferentes classes de ativo e calcule os log-returns diários, semanais e mensais da série.

Criando a lista de ações e capturando os preços:

```python
stocks = ['BRPR3.SA','ENBR3.SA','BVMF3.SA','PETR4.SA','VALE5.SA']

prices = pd.DataFrame()
for stock in stocks:
    s = Share(stock).get_historical('2000-01-01', '2015-12-31')
    s = pd.DataFrame(s).pivot_table(index = "Date", columns = "Symbol", values = "Adj_Close",aggfunc = "last")
    prices = pd.concat([prices, s],axis=1)

```

```python
prices.index = pd.to_datetime(prices.index)    
prices = prices["2006-01-01":]
prices = prices.replace(to_replace='NaN',value=np.nan)
prices = prices.fillna(method='ffill')
# prices = prices.fillna(method='bfill')
prices = prices.convert_objects(convert_numeric=True)
```

```python
dailyPrices = prices
dailyPrices.head()
```

```python
weeklyPrices = prices.resample(rule="W").last()
weeklyPrices.head()
```

```python
monthlyPrices = prices.resample(rule="M").last()
monthlyPrices.head()
```

```python
annualPrices = prices.resample(rule="A").last()
annualPrices.head()
```

```python
dailyReturns = dailyPrices.pct_change()
dailyReturns.head()
```

```python
weeklyReturns = weeklyPrices.pct_change()
weeklyReturns.head()
```

```python
monthlyReturns = monthlyPrices.pct_change()
monthlyReturns.head()
```

```python
annualReturns = annualPrices.pct_change()
annualReturns.head()
```

# Item: Calcule a FAC dos log-retornos diários, semanais e mensais e ajuste um modelo ARIMA


## Autocorrelação

```python
def plotAutoCorr(returns):
    for stock in returns.columns:
        plt.figure()
        plt.title(stock)
        autocorrelation_plot(returns[stock].dropna()) 
```

### Daily

```python
plotAutoCorr(dailyReturns)
```

### Weekly

```python
plotAutoCorr(weeklyReturns)
```

### Monthly

```python
plotAutoCorr(monthlyReturns)
```

Podemos observar que em quase todos os casos a autocorrelação não é estatísticamente diferente de zero. Exceto por alguns casos pontuais que não possuem interpretação lógica. Consideramos estes casos como aleatoriedade.

## ARIMA Model

```python
def estimateArimaForStocks(data,p=2,d=2,q=0):
    for stock in data.columns:
        print('\n\n\n\n Stocks: ' + stock + '\n\n')
        model = sm.tsa.arima_model.ARIMA(data[stock].dropna(), (p,d,q)).fit()
        print(model.summary())
```

### Daily

```python
estimateArimaForStocks(dailyReturns)
```

### Weekly

```python
estimateArimaForStocks(weeklyReturns)
```

### Monthly

```python
estimateArimaForStocks(monthlyReturns)
```

### Conclusão

Para os dados diários não encontramos o resultado esperado, pois utilizando um ARMA(2,2) os coeficientes dos primeiro e segundo lags apresentam significância para quase todas as ações.

Nos dados semanais é um pouco ambíguo, a maior parte das ações não demonstra estrutura de correlação.

Para dados mensais encontramos o resultado esperado no sentido de parecer não haver estrutura de correlação @ 1% significância.

# VAR

Precisamos criar a série de dividend yield

```python
def getDividends(stock,initialDate,finalDate):
    initialDate = datetime.datetime.strptime(initialDate,"%Y-%m-%d")
    finalDate = datetime.datetime.strptime(finalDate,"%Y-%m-%d")
    url="""http://ichart.yahoo.com/table.csv?s={stock}&c={initialYear!s}&a={initialMonth!s}&b={initialDay!s}&f={finalYear!s}&d={finalMonth!s}&e={finalDay!s}&g=v&ignore=.csv""".format(stock=stock,initialYear=initialDate.year,
                              initialMonth=initialDate.month,initialDay=initialDate.day,
                              finalYear=finalDate.year,finalMonth=finalDate.month,
                              finalDay=finalDate.day)
    csvStr = urllib.request.urlopen(url)
    df = pd.read_csv(csvStr)
#     df.pivot_table(index="Date", columns=stock, values="Dividends")
    df.set_index(keys='Date',inplace=True)
    df.columns = [stock]
    return df

```

```python
dividends = pd.DataFrame()
for stock in stocks:
    s = getDividends(stock,"2000-01-01","2015-12-31")
    dividends = pd.concat([dividends, s],axis=1)

dividends = dividends["2006-01-01":]
dividends.set_index(pd.to_datetime(dividends.index),inplace=True)
dividends.head()
```

```python
yearDividends = dividends.resample('A').sum()
yearDividends.plot()
```

Tivemos que mudar a frequência para anual pois as ações não pagam dividendos com muita frequência. Infelizmente isto torna nossa amostra pequena demais para esperarmos conclusões robustas.

```python
dividendYield = yearDividends / annualPrices
dividendYield.plot()
```

```python
import statsmodels.tsa.vector_ar.var_model as smvar
```

Por conta da discontinuidade dos dados iremos utilizar apenas as séries de ENBR3 e PETR4

## VAR ENBR3

```python
varData = pd.concat([pd.DataFrame(annualReturns['ENBR3.SA']), pd.DataFrame(dividendYield['ENBR3.SA'])], axis=1)
varData.columns = ['Returns','DividendYield']
varData = varData.fillna(0)
varData.head()
```

```python
model1 = smvar.VAR(varData)
results1 = model1.fit(1)
results1.summary()
```

O único coeficiente que possui significância a 1% no nosso VAR é a defasagem de DividendYield em Returns exatamente como esperávamos.

```python
model2 = smvar.VAR(varData)
results2 = model2.fit(2)
results2.summary()
```

# Função Resposta a Impulso

Como apenas o VAR de ENBR3 apresentou significância utilizaremos apenas ele.

Como temos apenas 11 pontos na amostra iremos fazer a resposta ao impulso 4 lags a frente.

Os impulsos são calculados em resposta a um choque de uma unidade em cada uma das variáveis.

```python
varData = pd.concat([pd.DataFrame(annualReturns['ENBR3.SA']), pd.DataFrame(dividendYield['ENBR3.SA'])], axis=1)
varData.columns = ['Returns','DividendYield']
varData = varData.fillna(0)
model1 = smvar.VAR(varData)
results1 = model1.fit(1)
```

```python
irf = results1.irf(4)
irf.plot(orth=False,impulse='DividendYield')
```

Podemos observar que no nosso modelo um choque no dividend yield gera um aumento no retorno esperado nos próximos 5 anos.
