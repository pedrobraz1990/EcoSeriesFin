# Nowcasting Brazilian Inflation Rate

In this essay we'll attempt to forecast brazilian inflation rate using big data and both PCA and partial squares techniques

# Big Data

On a first naive attemp to gather "big data" we'll simply check the most correlated, at least superficially sensical, variables brought up by google's ``correlate`` tool. [Searching for "Inflação", with no shift and for Brazil only](https://www.google.com/trends/correlate/search?e=inflação&t=weekly&p=br#default,30), we only find seemingly non sensical results, as the following examples:

* crítica
* hegel
* renal
* resistência
* adolescência
* distribuição
* diagrama
* características

And several others.

To avoid any spurious and purely coincidental results we proceed trying to shift the series 3 months (or approximately 12 weeks) because that is the frequency that inflation rates are published. With [this search](https://www.google.com/trends/correlate/search?e=inflação&t=weekly&shift=12&p=br#default,30), we verified the at least slightly credible results to be:

* universidade federal de
* universidade federal
* caes
* especializacao
* bikini
* cursos de ingles
* escolas de ingles
* mestrado em
* mestrado

It is worth mentioning that the most correlated variable, ``universidade federal de`` , correlation value is ``76.90%``, which we do not consider to be extremely high. Yet, we'll use it as a first trial.
